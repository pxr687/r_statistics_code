```r
# using a coinflip to decide if the samples are drawn from the same, or different, populations

coinflip = sample(c(0,1), 1)

population_1 = rgamma(n = 1000, 2)
population_2 = rgamma(n = 1000, 4)

if(coinflip == 1){samp_1 = sample(population_1, 100)
                  samp_2 = sample(population_2, 100)
                 } else {samp_1 = sample(population_1, 100)
                         samp_2 = sample(population_1, 100)}


# testing the samples for normality

print(shapiro.test(samp_1))
print(shapiro.test(samp_2))

res_1 = shapiro.test(samp_1)

res_2 = shapiro.test(samp_2)

# performing a Mann-Whitney U test if at least one sample is not normally distributed, else performing a t-test

if(res_1$p.value <= 0.05 | res_2$p.value <=0.5){wilcox.test(samp_1, samp_2)
                                               }else {t.test(samp_1,samp_2)}
```

```r
# plotting histograms of the samples (and qq plots for each sample)
hist(samp_1, col = 'blue')
qqnorm(samp_1, pch = 10, frame = FALSE)
qqline(samp_1, col = "blue", lwd = 2)

hist(samp_2, col = 'red')
qqnorm(samp_2, pch = 10, frame = FALSE)
qqline(samp_2, col = "red", lwd = 2)
```
